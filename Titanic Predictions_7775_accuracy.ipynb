```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#load data
data=pd.read_csv('titanic_train.csv')
data
```

```python
#check columns with missing values
data.isna().any()
```

```python
#data[data['Cabin'].isnull()]
#data['Name'].unique()
data[~data['Name'].str.contains("Mr. |Mrs.|Miss|Master")]["Age"].mean()
```

```python
#Finding mean value of ages based on Name column in training data

MR=data[data['Name'].str.contains("Mr. ")]["Age"].mean()
MRS=data[data['Name'].str.contains("Mrs.")]["Age"].mean()
MISS=data[data['Name'].str.contains("Miss")]["Age"].mean()
MASTER=data[data['Name'].str.contains("Master")]["Age"].mean()
REST=data[~data['Name'].str.contains("Mr. |Mrs.|Miss|Master")]["Age"].mean()
print(MR,MRS,MISS,MASTER,REST)
```

```python
#Removing columns with less significant information
y_train=data['Survived']
x_train=data.drop(['PassengerId','Survived','Ticket','Fare','Cabin','Embarked','SibSp'],axis=1)
x_train['Sex'].replace(['female','male'],[0,1],inplace=True)
x_train
```

```python
#Replace the Nan values in 'Age' column based on mean values from 'Name' column in training data

idx1 = np.where((np.isnan(x_train["Age"])) & (x_train['Name'].str.contains("Mr. ")))
temp1=x_train.loc[idx1]
temp1["Age"]=temp1["Age"].fillna(MR)
idx2=np.where((np.isnan(x_train["Age"])) & (x_train['Name'].str.contains("Miss")))
temp2=x_train.loc[idx2]
temp2["Age"]=temp2["Age"].fillna(MISS)
idx3 = np.where((np.isnan(x_train["Age"])) & (x_train['Name'].str.contains("Mrs. ")))
temp3=x_train.loc[idx3]
temp3["Age"]=temp3["Age"].fillna(MRS)
idx4 = np.where((np.isnan(x_train["Age"])) & (x_train['Name'].str.contains("Master")))
temp4=x_train.loc[idx4]
temp4["Age"]=temp4["Age"].fillna(MASTER)
idx5 = np.where((np.isnan(x_train["Age"])) & (~x_train['Name'].str.contains("Mr. |Mrs.|Miss|Master")))
temp5=x_train.loc[idx5]
temp5["Age"]=temp5["Age"].fillna(REST)
result=pd.concat([temp1,temp2,temp3,temp4,temp5])
result
```

```python
#Concat and Final preparation of training data

x_train=x_train.fillna(result)
x_train=x_train.drop(["Name"],axis=1)
x_train
```

```python
#Test data preparation
data2=pd.read_csv('titanic_test.csv')
x_test=data2.drop(['PassengerId','Ticket','Fare','Cabin','Embarked','SibSp'],axis=1)
x_test['Sex'].replace(['female','male'],[0,1],inplace=True)
x_test
```
